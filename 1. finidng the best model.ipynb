4.finding which model is best to use

```python
X = d.drop('cardio',axis=1)  
y = d[['cardio']]
```

```python
X.head(10)
```

```python
y.head(10)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,x_test,Y_train,y_test = train_test_split(X,y,random_state=356)
```

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.linear_model import LogisticRegression  # we cant use LinearRegression since the data consists of categorical
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
```

```python
lr = LogisticRegression()
kn = KNeighborsClassifier()
dt = DecisionTreeClassifier()
dr=DecisionTreeRegressor()
```

5.training the data with all classifiers

```python
lr.fit(X_train,Y_train)
kn.fit(X_train,Y_train)
dt.fit(X_train,Y_train)
dr.fit(X_train,Y_train)
```

## find which model is best
6.Take individual metrics to validate highest score with AUC in ROC

```python
lr.score(x_test,y_test)
```

```python
kn.score(x_test,y_test)
```

```python
dt.score(x_test,y_test)
```

```python
dr.score(x_test,y_test)
```

## Finally Plot ROC AUC curve to find the best model

```python
from sklearn.metrics import  plot_roc_curve
import warnings
```

```python
disp = plot_roc_curve(lr,x_test,y_test)
plot_roc_curve(kn,x_test,y_test,ax=disp.ax_)
plot_roc_curve(dt,x_test,y_test,ax=disp.ax_)
plot_roc_curve(dr,x_test,y_test,ax=disp.ax_)
```
